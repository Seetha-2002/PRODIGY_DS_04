<h1>Task 4</h1>

```python
import pandas as pd
import re
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
url = 'https://github.com/Seetha-2002/PRODIGY_DS_04/raw/main/twitter_training.csv'
cols=['ID', 'Topic', 'Sentiment', 'Text']
df= pd.read_csv(url,names=cols)
print(df.head())
```

```python
df.drop_duplicates(inplace=True)
```

<h2>Sentiment Distribution</h2>

```python
sns.countplot(x = 'Sentiment',data=df,palette='Set2')
plt.show()
```

```python
# counts for each sentiment
sent_counts = df['Sentiment'].value_counts()
sent_counts 
```

```python
# pie chart
plt.figure(figsize=(8, 8))
plt.pie(sent_counts, labels=sent_counts.index, autopct="%1.1f%%", startangle=140, colors=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Sentiment Distribution')
plt.show()
```

<h2>Analysis of Sentiment Distribution Across Various Topics</h2>

```python
plt.figure(figsize=(20,12))
sns.countplot(x='Topic',data=df,palette='Set1',hue='Sentiment')
plt.xticks(rotation=90)
plt.show()
```

<h2>Sentiment Distribution for Facebook</h2>

```python
fb_data = df[df['Topic'] == 'Facebook']
sent_counts = fb_data['Sentiment'].value_counts()

# pie chart
plt.figure(figsize=(8, 8))
plt.pie(sent_counts, labels=sent_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution of Topic "Facebook"')
plt.show()


#Word cloud for facebook(negative)

from wordcloud import WordCloud
topic = 'Facebook'
sentiment = 'Negative'
df['Text'] = df['Text'].astype(str)
text = " ".join(review for review in df[(df['Topic'] == topic) & (df['Sentiment'] == sentiment)]['Text'])
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title(f'Word Cloud for {sentiment} Sentiment in {topic}')
plt.show()
#positive
sentiment = 'Positive'  
text = " ".join(review for review in df[(df['Topic'] == topic) & (df['Sentiment'] == sentiment)]['Text'])
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title(f'Word Cloud for {sentiment} Sentiment in {topic}')
plt.show()
#neutral
sentiment = 'Neutral'  
text = " ".join(review for review in df[(df['Topic'] == topic) & (df['Sentiment'] == sentiment)]['Text'])
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title(f'Word Cloud for {sentiment} Sentiment in {topic}')
plt.show()
 
```
